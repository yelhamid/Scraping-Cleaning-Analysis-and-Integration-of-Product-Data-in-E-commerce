```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.preprocessing import normalize

# Assuming you have amazon_hk and aliexpress_hk DataFrames already defined

# Create a TfidfVectorizer to convert text data into TF-IDF vectors
vectorizer = TfidfVectorizer()

# Fit and transform the text data in both datasets for 'name'
tfidf_matrix1_name = vectorizer.fit_transform(concatenated_df['name'].fillna(''))
tfidf_matrix2_name = vectorizer.transform(concatenated_df_ali['name'].fillna(''))

# Normalize the TF-IDF matrices for 'name'
tfidf_matrix1_name_normalized = normalize(tfidf_matrix1_name, norm='l2', axis=1)
tfidf_matrix2_name_normalized = normalize(tfidf_matrix2_name, norm='l2', axis=1)

# Fit and transform the text data in both datasets for 'description'
tfidf_matrix1_description = vectorizer.fit_transform(concatenated_df['description'].fillna(''))
tfidf_matrix2_description = vectorizer.transform(concatenated_df_ali['description'].fillna(''))

# Normalize the TF-IDF matrices for 'description'
tfidf_matrix1_description_normalized = normalize(tfidf_matrix1_description, norm='l2', axis=1)
tfidf_matrix2_description_normalized = normalize(tfidf_matrix2_description, norm='l2', axis=1)

# Initialize an empty DataFrame to store the matching results
result_df_name = pd.DataFrame()
result_df_description = pd.DataFrame()

# Initialize lists to store similarity scores
name_similarity_scores = []
description_similarity_scores = []

# Define thresholds for 'name' and 'description'
name_threshold = 0.7  # Adjust as needed
description_threshold = 0.7 # Adjust as needed

# Iterate through rows in dataset1 and find the most similar row in dataset2 for 'name'
for i in range(len(concatenated_df)):
    row1_name = tfidf_matrix1_name_normalized[i]
    similarity_scores_name = cosine_similarity(row1_name, tfidf_matrix2_name_normalized).flatten()

    # Find the index of the maximum similarity score for 'name'
    max_similarity_index_name = similarity_scores_name.argmax()

    # Append the similarity score for 'name' to the list
    name_similarity_scores.append(similarity_scores_name[max_similarity_index_name])

    # Add the matching rows for 'name' to the result DataFrame
    for col in concatenated_df.columns:
        result_df_name.at[i, f"{col}_amazon_hk"] = concatenated_df.at[i, col]
        result_df_name.at[i, f"{col}_aliexpress_hk"] = concatenated_df_ali.at[max_similarity_index_name, col]

# Add the similarity score column for 'name' to the result DataFrame
result_df_name['name_similarity_score'] = name_similarity_scores

# Filter the results based on the respective similarity threshold for 'name'
result_df_name = result_df_name[
    (result_df_name['name_similarity_score'] > name_threshold) & 
    (result_df_name['name_similarity_score'] > 0.5)]
# Iterate through rows in dataset1 and find the most similar row in dataset2 for 'description'
for i in range(len(amazon_hk)):
    row1_description = tfidf_matrix1_description_normalized[i]
    similarity_scores_description = cosine_similarity(row1_description, tfidf_matrix2_description_normalized).flatten()

    # Find the index of the maximum similarity score for 'description'
    max_similarity_index_description = similarity_scores_description.argmax()

    # Append the similarity score for 'description' to the list
    description_similarity_scores.append(similarity_scores_description[max_similarity_index_description])

    # Add the matching rows for 'description' to the result DataFrame
    for col in concatenated_df.columns:
        result_df_description.at[i, f"{col}_amazon_hk"] = concatenated_df.at[i, col]
        result_df_description.at[i, f"{col}_aliexpress_hk"] = concatenated_df_ali.at[max_similarity_index_description, col]

# Add the similarity score column for 'description' to the result DataFrame
result_df_description['description_similarity_score'] = description_similarity_scores

# Filter the results based on the respective similarity threshold for 'description'
result_df_description = result_df_description[result_df_description['description_similarity_score'] > description_threshold]
```

```python
# Assuming you have result_df_name and result_df_description DataFrames from the previous code

# Merge result_df_name and result_df_description on common columns
result_df_combined = pd.merge(
    result_df_name,
    result_df_description,
    how='inner',
    left_on=['id_amazon_hk', 'link_amazon_hk'],
    right_on=['id_amazon_hk', 'link_amazon_hk'],
    suffixes=('_name', '_description')
)

# Combine the similarity scores for 'name' and 'description'
result_df_combined['combined_similarity_score'] = (
    result_df_combined['name_similarity_score'] +
    result_df_combined['description_similarity_score']
)

# Filter the results based on the combined similarity threshold
combined_threshold = 0.3  # Adjust as needed
result_df_combined = result_df_combined[
    (result_df_combined['combined_similarity_score'] > combined_threshold)
]

# Display the resulting DataFrame
print(result_df_combined)
```
